```python
df=df.drop(['Customer_ID', 'Age', 'Gender','Email_Opt_In', 'Promotion_Response'],axis=1)
df.head()
```

```python
df['Target_Churn'] = df['Target_Churn'].replace({'True': 1, 'False': 0})
```

```python
df.head()
```

```python
m=MinMaxScaler()
df['Total_Spend']=m.fit_transform(df[['Total_Spend']])
df['Annual_Income']=m.fit_transform(df[['Annual_Income']])
df['Average_Transaction_Amount']=m.fit_transform(df[['Average_Transaction_Amount']])
df['Last_Purchase_Days_Ago']=m.fit_transform(df[['Last_Purchase_Days_Ago']])
df.head()
```

```python

df['Target_Churn'] = df['Target_Churn'].replace({'True': 1, 'False': 0})
```

```python
x=df.drop('Target_Churn',axis='columns')
y=df.Target_Churn
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
