```python
col=['Diet',
 'Heating Energy Source',
 'Transport',
 'Vehicle Type',
 'Frequency of Traveling by Air',
 'Energy efficiency',
 'Recycling',
 'Cooking_With']
df1=pd.get_dummies(df,columns=col,dtype=int,drop_first=True)
```

```python
x=df1.drop('CarbonEmission',axis=1)
y=df1.CarbonEmission
len(x),len(y)
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
len(x_train),len(x_test)
```

```python
s=cross_val_score(LinearRegression(),x_train,y_train)
s1=cross_val_score(DecisionTreeRegressor(),x_train,y_train)
s2=cross_val_score(RandomForestRegressor(n_estimators=100),x_train,y_train)
np.average(s),np.average(s1),np.average(s2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_train,y_train),model.score(x_test,y_test)
```
