```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
df=pd.read_csv("boston.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# sc=MinMaxScaler()
sc=StandardScaler()
cols=['INDUS','RM','AGE','DIS','TAX','PTRATIO','B','LSTAT']
df[['INDUS','RM','AGE','DIS','TAX','PTRATIO','B','LSTAT']]=sc.fit_transform(df[['INDUS','RM','AGE','DIS','TAX','PTRATIO','B','LSTAT']])
df
```

```python
x=df.drop(['MEDV','ZN'],axis='columns')
y=df.MEDV
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
s=cross_val_score(LinearRegression(),x_train,y_train)
np.average(s)
```

```python
s1=cross_val_score(DecisionTreeRegressor(),x_train,y_train)
np.average(s1)
```

```python
s2=cross_val_score(RandomForestRegressor(n_estimators=5),x_train,y_train)
np.average(s1)
```

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
clf=GridSearchCV(RandomForestRegressor(), {
        'n_estimators':[1,5,10],
    }, 
    cv=5, 
    return_train_score=False, 

)
clf.fit(x_train,y_train)
# clf.best_score_
clf.best_params_
```

```python
model=RandomForestRegressor(n_estimators=1000,max_depth=10,              # Limit tree depth
    min_samples_split=10,      # Minimum samples required to split a node
    min_samples_leaf=4,  )
# model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
model.score(x_test,y_test)
```
